### .loc, Um Método Muito Importante

    - Pegar 1 linha
    - Pegar linhas de acordo com alguma condição
    - Pegar linhas e colunas específicas
    - Pegar 1 valor específico

```python
# Pegar uma linha : 
display(venda_df.loc[1:5])

# Pegar linhas de acordo com as condições : 
vendas_norteshopping_df = venda_df.loc[venda_df['ID Loja'] == 'Norte Shopping']

# Pegar várias linhas e colunas usando o loc : 
venda_df = pd.read_excel("f:/DaLucas_Codes/PYTHON FUNDAMENTALS/ARQUIVOS EXCEL/Vendas.xlsx")
vendas_norteshopping_df = venda_df.loc[venda_df['ID Loja'] == 'Norte Shopping', ["ID Loja", "Produto", "Quantidade"]]
display(vendas_norteshopping_df)

# Pegar 1 valor específico : 
display(venda_df.loc[1, 'Produto'])
```

### Adicionar 1 coluna

```python
# A partir de uma coluna que existe : 
venda_df['Comissão'] = venda_df['Valor Final'] * 0.05
display(venda_df)

# Criar uma coluna com valor padrão : 
venda_df.loc[:, "Imposto"] = 0
display(venda_df)
```

### Adicionar 1 linha

- Linhas de um complemento da base de dados

```python
import pandas as pd

vendas_dez_df = pd.read_excel("f:/DaLucas_Codes/PYTHON FUNDAMENTALS/ARQUIVOS EXCEL/Vendas - Dez.xlsx")
# display(vendas_dez_df)

# Substituindo append por concat pois a nova versão do pandas não permite usar .append
venda_df = pd.concat([venda_df, vendas_dez_df])
display(venda_df)
```

### Excluir linhas e colunas

```python
venda_df = venda_df.drop("Imposto", axis=1)
display(venda_df)
```

### Calcular Indicadores

- Groupby
- Value Counts

```python
# Value Counts : 
transacoes_loja = venda_df['ID Loja'].value_counts()
display(transacoes_loja)

# Group by : 
faturamento_produto = venda_df[['Produto', 'Valor Final']].groupby('Produto').sum()
display(faturamento_produto)
```
