```python
carro = float(input("Digite o valor do carro: "))
entrada = float(input("Digite o valor da entrada: "))
saldo = float(input("Digite o seu saldo: "))
parcelas = int(input("Digite a quantidade de parcelas: "))

restante = carro - entrada
parcela = restante / parcelas
print(f"A sua parcela será: R${parcela:.2f}")

if entrada > saldo:
    print("Saldo Insuficiente")
else:
    print("Simulação Aprovada")
```
