### Contador de Palavras

```python
import pandas as pd

tabela = pd.read_csv("f:/DaLucas_Codes/PYTHON FUNDAMENTALS/Arquivos Csv/ReviewsIphoneX.csv")
display(tabela)
```

```python
texto = " ".join(tabela["body"].astype(str))
texto = texto.lower()

import re

lista_palavras = re.findall(r"\b\w+\b", texto)

from collections import Counter

contagem = Counter(lista_palavras)
print(contagem)

for item in contagem.most_common(200):
    print(item)
```
