### Sempre que for criar uma janela começe com : 

```python
janela = Tk() # Abre a janela

janela.mainloop() # Não deixa a janela feixar
```

### Mudar o Título : 


```python
janela.title("Cotação Atual das Moedas")
```

### Colocar texto : 


```python
texto_orientacao = Label(janela, text="Clique no botão para ver as cotações atuais das moedas ")
```

### A onde coloca ?

```python
texto_orientacao.grid(column=0, row=0, padx=10, pady=10) # coluna 0, linha 0, espaço do eixo X (Horizontal (Deitado)), espaço do eixo Y (Vertical (Em Pé))
```
